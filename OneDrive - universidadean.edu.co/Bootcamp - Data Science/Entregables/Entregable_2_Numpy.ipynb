# 1. Crea un arreglo 5x3 de números aleatorios entre 0 y 100

```python
import numpy as np
arreglo_aleatorio = np.random.randint(0, 101, size=(5, 3))
arreglo_aleatorio
```

# 2. ¿Cuál es la suma total de los valores del arreglo?

```python
print(f"Sumatoria de los valores del arreglo: {np.sum(arreglo_aleatorio)}")
```

# 3. ¿Cúal es el promedio de cada columna del arreglo?

```python
print(f"Promedio de los valores de cada columna del arreglo: {np.std(arreglo_aleatorio, axis=0)}")
```

# 6. Abre una imagen, muestra shape, min, max y promedio de cada canal

```python
from PIL import Image
import requests

img = Image.open(requests.get("https://media.es.wired.com/photos/657cb5b81e17b099f8f9e15c/3:2/w_3009,h_2006,c_limit/gatos%20172050389.jpg", stream=True).raw)
img_array = np.array(img)

import matplotlib.pyplot as plt
plt.imshow(img_array)
plt.axis('off')
plt.show()
```

```python
print(f"Shape de imagen: {img_array.shape}")
```

```python
# Separar los canales
red_channel = img_array[:, :, 0]
green_channel = img_array[:, :, 1]
blue_channel = img_array[:, :, 2]

# Imprimir valores de cada canal
print(f"\nCanal rojo:")
print(f"  Min: {np.min(red_channel)}")
print(f"  Max: {np.max(red_channel)}")
print(f"  Average: {np.mean(red_channel)}")

print(f"\nCanal verde:")
print(f"  Min: {np.min(green_channel)}")
print(f"  Max: {np.max(green_channel)}")
print(f"  Average: {np.mean(green_channel)}")

print(f"\nCanal azul:")
print(f"  Min: {np.min(blue_channel)}")
print(f"  Max: {np.max(blue_channel)}")
print(f"  Average: {np.mean(blue_channel)}")
```
