# 1. Leer un archivo del formato de su elección sobre cualquier temática

```python
import pandas as pd
database = pd.read_csv(ruta_archivo, encoding='latin-1', sep=";",on_bad_lines='skip')  # Leer el dataset en formato CSV


#Se usa encoding latin-1 ya que la base de datos fue guardada en esa condificación.
#Se usa skip en on_bad_lines dado que una línea sólo contenia 2 columnas y generaba error al importar.
```

#3. Genere un nuevo dataset aplicando un filtro sobre los datos originales

```python
print("Bases de datos tipo CSV y que su última actualización sea después del 2020")
database_filtrado = database[(database['Data_Type'] == "CSV") & (database['Last_Update'] > 2020)]
database_filtrado.head()
```

#4. Genere un nuevo dataset aplicando un ordenamiento sobre los datos originales

```python
#Como se evidencia que los tamaños de los archivos están en diferentes unidades, se estandariza a KB.

def convert_size_to_kb(size):
  """Convierte el tamaño como '100MB', '2GB' a kilobytes."""
  size = size.strip().upper()
  if 'KB' in size:
    return float(size.replace('KB', ''))
  elif 'MB' in size:
    return float(size.replace('MB', '')) * 1024
  elif 'GB' in size:
    return float(size.replace('GB', '')) * 1024 * 1024
  else:
    return None

database['Size_KB'] = database['Size'].apply(convert_size_to_kb)


print("\nDataset con Size en KB:")
print(database[['Size', 'Size_KB']].head())
```

```python
print("Ordenar el tamaño de los datasets por orden descendente")
database.sort_values(by='Size_KB', ascending=False)
```
